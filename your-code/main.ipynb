#### Question 1: Find the questions and answers of last month.

```python
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=1591488000, todate=1594080000)
questions
```

```python
from pandas.io.json import json_normalize
data_q = json_normalize(questions["items"])
data_q
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
#we get the most voted questions from today
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
questions_t = SITE.fetch('questions', fromdate=1594080000, todate=1594135380, tagged='python', sort='votes', min=5)
questions_t
```

```python
#we convert the most voted questions to dataframe
from pandas.io.json import json_normalize
data_today = json_normalize(questions_t["items"])
data_today
```
